```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = load_diabetes()
df
```

```python
data = pd.DataFrame(df.data)
data
```

```python
data.columns = df.feature_names
data
```

```python
#diving data into dependant and independant features

x = data #independant
y = df.target #dependant

# df["target"]
# df.target both are basically the same but use first one for safety, clarity, and consistency.
```

```python
#train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)

# We imported train_test_split from sklearn.model_selection
# to split our dataset into training and testing parts.
# In the line x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0),
# we're using that function to divide our data — where x is the input (features) and y is the output (target).
# The test_size=0.2 means 20% of the data will be used for testing, and the remaining 80% will be used for
# training the model. The random_state=0 just makes sure that the data split is always the same every
# time we run the code. Even if we run it today or after a week, the train and test sets will stay
# consistent. This is useful for debugging, comparing models fairly, and getting repeatable results.
# So even if we only run the code once, it’s a good practice to add random_state to keep things stable.

```

```python
#standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
# We use the scaler function to normalize values — for example, if the "Age" column has values like 22, 25, 30
# which are far apart or random, then we bring them into a specific range like -1 to +1. This helps the machine treat all features equally.

# Otherwise, the machine might wrongly assume that a higher age always means higher salary just because the numbers are bigger.

# So, we use scaler.fit() to calculate the mean and standard deviation of the data,
# and then scaler.transform() to replace each original value with its normalized version using the formula:

# z = (x - mean) / std
# For example:
# If age was 22, and mean = 25, std = 2, then:


# z = (22 - 25) / 2 = -1.5
# So the original age 22 is replaced with -1.5.


x_test = scaler.transform(x_test)

# #So we applied fit() on training data so that the machine could calculate the mean and standard deviation.
# Then, we used transform() to apply the formula on the training values and normalize them.
# Later, for the test data, we used the same mean and std that were already stored during training —
# and simply used transform() on test data to scale it using those same training values.
# No need to re-fit on test — just apply the trained transformation.


#Okay, so fit calculates the mean and standard deviation,
#  and transform uses those values to apply the formula
#  and actually normalize (i.e., convert) the data

#Generally we dont transform the output data

#So in step (test train split), we split the data into training and testing sets, but this split is actually used later in
#  step 5 (the model training) during the training process, And standardization is also applied during step 5,
```

```python
#applying linear regression

from sklearn.linear_model import LinearRegression

#cross validation
from sklearn.model_selection import cross_val_score
```

```python
#model train

reg = LinearRegression()
reg.fit(x_train, y_train)

mse= cross_val_score(reg,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
#first argument is model then input then output
# mse
np.mean(mse)

#.fit() performs different tasks depending on the object it’s used with.


#  For example, StandardScaler().fit() calculates the mean and standard
#  deviation of the data for scaling, while LinearRegression().fit() is
#   used to teach the model the pattern between the input and output data.
#    The scoring parameter defines which metric will be used to evaluate
#     the model’s performance — basically, how the error is calculated.
#     In our case, we used neg_mean_squared_error, which refers to the
#      negative of the mean squared error: (predicted − actual)². This value represents how far off
#  the model’s predictions are from the actual results — and the smaller
#   this error, the better the model is performing.
```

```python
#Prediction
predictions = reg.predict(x_test)
predictions #now checkt the predictions are correct so well compare with the y_test)
```

```python
import seaborn as sns
sns.distplot(y_test-predictions)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,predictions)

#values between 0 -1 1 is model predicted very accurately
# R² score basically tells how well the model has predicted
# how much of the actual output variance is explained by the model.
#  In r2_score(y_test, predictions), the first argument is the actual values
#  , and the second is the model's predicted output. So if the result is 0.33,
#   it means the model explained 33% of the variance, and 67% remains as error.
#    During training, we often use Mean Squared Error (MSE) — or neg_mean_squared_error
#     to help the model learn by minimizing the difference between predicted and actual values
#     . It's useful in training because it directly guides the optimization. However,
#      once training is done and the model starts making real predictions, R² is recommended for evaluation,
#       because it gives a more interpretable metric of performance.
# So yes — MSE is best for training, and R² is best for validating model performance after prediction.
```
